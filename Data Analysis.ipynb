### Here are the primary features to visualise:

1. LotArea: Lot size in square feet
2. Street: Type of road access
3. Neighborhood: Physical locations within Ames city limits
4. OverallCond: Overall condition rating
5. YearBuilt: Original construction date
6. FullBath: Full bathrooms above grade
7. HalfBath: Half baths above grade
8. Bedroom: Number of bedrooms above basement level
9. Kitchen: Number of kitchens

```python
def plot_features(cols):
    %matplotlib inline
    if cols in categorical_cols:
        fig = plt.figure(figsize = (24,10))
        plt.subplot(1,2,1)
        plt.hist(train[cols], color= 'green')
        plt.ylabel('Counts')
        plt.xlabel(cols)
        plt.xticks(rotation = 90)
        plt.subplot(1,2,2)
        g = sns.boxplot(x = train[cols], y = dataset['SalePrice'])
        g.set_xticklabels(rotation = 90, labels= train[cols]) 
        plt.show()
    else:
        fig = plt.figure(figsize = (12,10))
        plt.scatter(x = train[cols], y = dataset['SalePrice'], c ='green')
        plt.ylabel('SalePrice of Home')
        plt.xlabel(cols)
    
```
